##### Introduction:

Climate transformations, such as rising temperatures, shifting precipitation patterns, and increasing extreme climatic events, are reshaping how landscape function and are influenced by human activities. While the impacts of climate change are global, they manifest differently across places due to varying environmental and cultural contexts. Climatic trends and impacts, shaped by specific cultural and environmental contexts, are understood and experienced differently by individuals and communities, carrying varied meanings over time. While science-based technologies with regional, long term data collection, and engineering are essential for characterizing climate change effects, they cannot fully capture the nuances and contingencies that emerge when these changes intersect with human lives. Many communities lack the required knowledge or technology to access this information which is directly influencing their everyday life.

This data pipeline project is offering a space for me to exercise the process of data engineering as a landscape designer, to explore the integration of multiple datasets to visualize climate trend and pattern over time at selected areas of interest. Through collecting and visualizing historical climatic data, remote sensing data, and census data from open sources, the final dashboard aims to set up a site background profile and help communities to understand climate change and impact through accessible visual and communication at selected sites.

##### Background:

Published Articles:

1. Overpeck, J.T., Meehl, G.A., Bony, S., & Easterling, D.R. (2011). Climate data challenges in the 21st century. Science, 331(6018), 700-702. https://doi.org/10.1126/science.1197869

    The authors emphasize that both natural and human-induced climate variability and change are having growing effects on both human and natural systems. To understand how the climate system functions, its past behavior, and potential future patterns, it is essential to rely on climate data, including instrumental and proxy observations, satellite records, and model outputs. Meanwhile, the amount, diversity, and complexity of climate data are exploding, as is the demand for this data from a wide range of users, from researchers and resource managers to the general public. Therefore, the challenge of accessing, understanding, formatting, and contextualizing climate data for non-experts becomes more obvious. The authors addresses the need for a more open, user-friendly data access paradigm.

    This paper presents the design and implementation of the Hawai'i Climate Data Portal which is a web-based platform for the automation, visualization, and dissemination of climatological data, focusing primarily on precipitation. Although using a more complex and advanced data engineering process, the research validated a rigorous data visual is highly useful for resource planning, preparation, as well as stakeholder engagement. The system automates the daily collection, processing of climate data, and quality control from multiple sensor networks and repositories, producing high-resolution gridded maps of rainfall and temperature. This example demonstrates and gives insight on the possibility of translating collected climate data to a spatially mapped interface for future interpretation.

Public Datasets:

1.  NOAA Climate API - National Centers for Environmental Information. (n.d.). Web services API: Version 2 Documentation. US Department of Commerce. Retrieved October 10, 2025, from https://www.ncdc.noaa.gov/cdo-web/webservices/v2

    NCDC's Climate Data Online offers web services that provide access to current climate data. Developers can use the database of weather and climate data with their own scripts or programs. An access token is required to use the API, and each token will be limited to five requests per second and 10,000 requests per day. (Introduction from webpage)

2. Open-Meteo - Open-Meteo. (n.d.). Free Weather API: Open-source weather forecasts worldwide. Retrieved October 11, 2025, from https://open-meteo.com/

    This is an open-source weather API and offers free access for non-commercial use. No API key required. It partners with national weather services and provide high resolution open data, ranging from 1 to 11 km. It's a JSON API. Some datasets are in global scale. (Introduction from webpage)

3. Sentinel Hub - Copernicus Data Space Ecosystem. (n.d.). Sentinel Hub. Retrieved October 10, 2025, from https://dataspace.copernicus.eu/analyse/apis/sentinel-hub

    This is a set of RESTful APIs that provide access to various satellite imagery archives. It allows users to access raw satellite data, rendered images, statistical analysis, and other features. The Sentinel Hub has the ability to automatically archive, process in real-time, and distribute remote sensing data and related Earth Observation products. Users can retrieve satellite data and process it for specific area of interest and time range utilizing APIs. (Introduction from webpage)

4. US Census Bureau - U.S. Census Bureau. (n.d.). Available APIs - Data sets. Retrieved October 12, 2025, from https://www.census.gov/data/developers/data-sets.html

    These are all the publicly available datasets of census data provided by the US Census Bureau. Data can be accessed in html, xml, and json formats. Many available datasets such as economic census, population estimates and projections, geography information, and poverty statistics are useful to analyze demographic and socio-economic status of a place.

##### Data:
1. NASA Goddard Space Flight Center. (n.d.). _DONKI: Data of NASA's Space Weather Research Center_. NASA. Retrieved from https://ccmc.gsfc.nasa.gov/tools/DONKI/

    The Space Weather Database of Notifications, Knowledge, Information (DONKI) provides daily updates on the interpretation of space weather observations, analyses, models, forecasts, and alerts from the Space Weather Research Center. The data source shares data through webservice API. There is a request limit of 1,000 requests per hour using NASA's API. API key can be generated with basic information (first name, last name and email). No specific licensing information found on the webpage.

2. Open-Meteo. (n.d.). _Climate API documentation_. Retrieved from https://open-meteo.com/en/docs/climate-api
		
    This API provides access to climate data from the Intergovernmental Panel on Climate Change (IPCC). Climate change is not a phenomenon that affects all regions equally. While some areas might experience more frequent and intense rainfall, others are confronted with extended periods of drought, which can have severe impact on agriculture. The change in temperature patterns is also unevenly distributed, with cities being particularly vulnerable to heatwaves. The API provides daily climate data rather than climate indices for more granular detail of changes. The data provides CSV file as well as API URL. It is licensed under a Creative Commons Attribution 4.0 International License.

3. Copernicus Data Space Ecosystem. (n.d.). _Sentinel Hub Process API documentation_. Retrieved from https://documentation.dataspace.copernicus.eu/APIs/SentinelHub/Process.html
		
    The Processing API is the most commonly used API in Sentinel Hub as it provides images based on satellite data. Raw satellite data, simple band combinations, and calculations of simple remote sensing indices can be requested from this API. Users can specify area of interest for a request. Data from Copernicus has a free, full and open data policy for non-commercial use.

4. Copernicus Data Space Ecosystem. (n.d.). Sentinel Hub Statistical API documentation. Retrieved from https://documentation.dataspace.copernicus.eu/APIs/SentinelHub/Statistical.html
		
    The Statistical API provides statistics calculated based on satellite imagery without having to download images. Example requests can be cloudy pixels for a given area of interest and time period. This API is similar to the Processing API but returning a calculated results in JSON format instead of images. Histograms can be requested too. Data from Copernicus has a free, full and open data policy for non-commercial use.

5. U.S. Census Bureau. (n.d.). _Economic Census_. U.S. Department of Commerce. Retrieved from https://www.census.gov/data/developers/data-sets/economic-census.html
		
    The Economic Census is the official measure of the Nation's businesses and economy. Held every five years, this survey acts as the statistical foundation for tracking current economic indicators, including the Gross Domestic Product and the Producer Price Index. This dataset has scale up to consolidated city which provide more granular details than state or county level. Accessing through API, there is no specific usage restrictions specified in this public domain. 

##### Potential Analyses:
Most of the data sources listed above include a time parameter that historical data can be extracted, engineered, and visualize. They feature climate patterns over time such as temperature, precipitation, wind direction, snow depth changes, etc. To visualize climate trend at area of interest on a public-facing dashboard, analyses can be histograms or line charts showing statistic over time. Extracted demographic and socio-economic contexts from census data can be shown as cross-tabs with featuring climate data to show correlation of climate change and community impacts.
